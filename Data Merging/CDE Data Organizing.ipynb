I see that not all the districts have a score, so I'm filtering out any rows that have "- -" as the score since this is missing. I'll do the same below, before re-combining the dataframes to have both the ELA and Math scores in the same dataframe.

Here I am changing some of the columns from object type to float type. I am also converting some of the numbers from strings to floats by removing punctuation then converting. This will get all of my variables in a better form for machine learning.

```python
cols = ['Attendance Rate', 'Truancy Rate', 'Percent Minority', 'Percent Female', 'Percent Male', 'Percent Non-Binary', 'Staff Turnover Rate']
merged_df9[cols] = merged_df9[cols].apply(lambda x: x.str.rstrip('%').astype(float) / 100)
```

```python
merged_df9['Student Teacher Ratio'] = merged_df9['PK-12 Count'] / merged_df9['Teacher FTE']
```

```python
merged_df9['Average Teacher Salary'] = merged_df9['Average Teacher Salary'].str.replace(',', '').astype(float)
```

```python
merged_df9['Federal Funding Per Pupil'] = merged_df9['Federal Funding Per Pupil'].str.replace(',', '').astype(float)
```

```python
merged_df9['State/Local Funding Per Pupil'] = merged_df9['State/Local Funding Per Pupil'].str.replace(',', '').astype(float)
```

```python
merged_df9['Total Funding Per Pupil'] = merged_df9['Federal Funding Per Pupil'] + merged_df9['State/Local Funding Per Pupil']
```
